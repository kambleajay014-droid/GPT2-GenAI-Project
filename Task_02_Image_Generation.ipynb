```python
# ===============================================
# ðŸ§  Task 02 - Image Generation with Pre-trained Models
# Using Stable Diffusion (Hugging Face diffusers)
# ===============================================

# STEP 1: Install dependencies
!pip install diffusers transformers accelerate safetensors torch --quiet

# STEP 2: Import required libraries
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import os

# STEP 3: Check device (GPU/CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"âœ… Using device: {device}")

# STEP 4: Load pre-trained Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16 if device=="cuda" else torch.float32)
pipe = pipe.to(device)

print("âœ… Model loaded successfully!")

# STEP 5: Generate an image from text prompt
prompt = "Dog wearing glasses and running on street"
print(f"ðŸŽ¨ Generating image for prompt: '{prompt}'")

image = pipe(prompt).images[0]

# STEP 6: Display the image
image.show()

# STEP 7: Save the image
output_path = "/content/generated_image.png"
image.save(output_path)
print(f"ðŸ’¾ Image saved at: {output_path}")
```
